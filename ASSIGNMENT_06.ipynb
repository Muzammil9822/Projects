```python
df = pd.DataFrame(texts)
df.columns = ['Text']
#Set the display option to show the entire content of the 'Paragraph' column
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
df
```

```python
# Get the embeddings of each text and add to an embeddings column in the dataframe
def embed_fn(text):
  return genai.generate_embeddings(model=model, text=text)['embedding']

df['Embeddings'] = df['Text'].apply(embed_fn)
df
```

# Document search with Q & A

```python
query = "Explain climate policies"
query
```

```python
def find_best_passage(query, dataframe):
  """
  Compute the distances between the query and each document in the dataframe
  using the dot product.
  """
  query_embedding = genai.generate_embeddings(model=model, text=query)
  dot_products = np.dot(np.stack(dataframe['Embeddings']), query_embedding['embedding'])
  idx = np.argmax(dot_products)
  return dataframe.iloc[idx]['Text'] # Return text from index with max value
```

```python
passage = find_best_passage(query, df)
passage
```

# Question and Answering Application

```python
def make_prompt(query, relevant_passage):
  escaped = relevant_passage.replace("'", "").replace('"', "").replace("\n", " ")
  prompt = textwrap.dedent("""You are a helpful and informative bot that answers questions using text from the reference passage included below. \
  Be sure to respond in a complete sentence, being comprehensive, including all relevant background information. \
  However, you are talking to a non-technical audience, so be sure to break down complicated concepts and \
  strike a friendly and converstional tone. \
  If the passage is irrelevant to the answer, you may ignore it.
  QUESTION: '{query}'
  PASSAGE: '{relevant_passage}'

    ANSWER:
  """).format(query=query, relevant_passage=escaped)

  return prompt
```

```python
prompt = make_prompt(query, passage)
print(prompt)
```

```python
text_models = [m for m in genai.list_models() if 'generateText' in m.supported_generation_methods]

text_model = text_models[0]
text_model
```

```python
temperature = 0.5
answer = genai.generate_text(prompt=prompt,
                            model=text_model,
                            candidate_count=3,
                            temperature=temperature,
                            max_output_tokens=1000)
```

```python
for i, candidate in enumerate(answer.candidates):
  print(f"Candidate {i}: {candidate['output']}\n")
```
